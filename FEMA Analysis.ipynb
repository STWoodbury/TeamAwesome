```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read in fema data
femadata = pd.read_csv('Datasets/DisasterDeclarationsSummaries.csv')
```

```python
#filter only to hurricane incidents
femadata = femadata[femadata['incidentType'] == "Hurricane"]
```

```python
#filter to years equal or greater than 1991
femadata = femadata[femadata['fyDeclared'] >= 1991]
```

```python
#get unique storms which triggered a disaster declaration for each year
disasterdeclarations = pd.DataFrame(femadata.groupby(['fyDeclared'])['disasterNumber'].nunique())
```

```python
year = disasterdeclarations.index
count = list(disasterdeclarations['disasterNumber'])
```

```python
plt.plot(year, count, marker="o")
plt.axhline(y=np.nanmean(count), color='red', linestyle='--', linewidth=2, label='Avg')
plt.title("Hurricane-Related National FEMA Disaster Declarations (1991-2022)")
plt.xlabel("Year")
plt.ylabel("Hurricane Disaster Declarations")
plt.savefig('Visualizations/FemaDisasterDeclarationsvsAverage')
```
