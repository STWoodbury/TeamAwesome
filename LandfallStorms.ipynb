```python
import pandas as pd
import matplotlib.pyplot
%matplotlib inline
```

```python
df = pd.read_csv('Datasets/PerfectStorm.csv')
```

```python
landfall = []
for index, row in df.iterrows():
    if row['Landfall Counts'] != 0:
        landfall.append('Landfall')
    else: 
        landfall.append('No Landfall')
df['Landfall Boolean'] = landfall
```

```python
df = df[df['Year'] >= 1991]
```

```python
df
```

```python
ax = df.groupby(['Year', 'Landfall Boolean']).size().unstack().plot(kind='bar', stacked=True, figsize=(18,5), width=.7);
for container in ax.containers:
    ax.bar_label(container)
```
